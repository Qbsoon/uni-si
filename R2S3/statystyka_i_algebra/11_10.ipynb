```python
import math

def add(x1,x2):
    if len(x1)!=len(x2):
        print(f'Różne długości wektorów: {len(x1)} i {len(x2)}')
        return "error"
    if x1[0]!=x2[0]:
        print(f'Wektory mają różną orientację')
        return "error"
    x=[x1[0]]
    for i in range(1,len(x1)):
        x.append(x1[i]+x2[i])
    return x

def min(x1,x2):
    if len(x1)!=len(x2):
        print(f'Różne długości wektorów: {len(x1)} i {len(x2)}')
        return "error"
    if x1[0]!=x2[0]:
        print(f'Wektory mają różną orientację')
        return "error"
    x=[x1[0]]
    for i in range(1,len(x1)):
        x.append(x1[i]-x2[i])
    return x

def multi_by_skalar(x1,x2):
    x=[x1[0]]
    for i in range(1,len(x1)):
        x.append(x1[i]*x2)
    return x

def multi_skalar(x1,x2):
    if len(x1)!=len(x2):
        print(f'Różne długości wektorów: {len(x1)} i {len(x2)}')
        return "error"
    if x1[0]!=x2[0]:
        print(f'Wektory mają różną orientację')
        return "error"
    x=0
    for i in range(1,len(x1)):
        x=x+(x1[i]*x2[i])
    return(x)

def multi_hadamard(x1,x2):
    if len(x1)!=len(x2):
        print(f'Różne długości wektorów: {len(x1)} i {len(x2)}')
        return "error"
    if x1[0]!=x2[0]:
        print(f'Wektory mają różną orientację')
        return "error"
    x=[x1[0]]
    for i in range(1,len(x1)):
        x.append(x1[i]*x2[i])
    return(x)

def multi_outer(x1,x2):
    if x1[0]==x2[0]:
        print(f'Wektory mają tą samą orientację')
        return "error"
    x=[]
    if x1[0]==0:
        for y in range(1,len(x2)):
            temp=[]
            for z in range(1,len(x1)):
                temp.append(x1[z]*x2[y])
            x.append(temp)
    return(x)

def mod(x1):
    x=0
    for i in range(1,len(x1)):
        x=x+(x1[i]*x1[i])
    return(math.sqrt(x))

def singular(x11):
    s=1/mod(x11)
    return multi_by_skalar(x11,s)

def tpoz(x1):
    if x1[0]==0:
        x1[0]=1
    else:
        x1[0]=0
    return(x1)

def com_line(x1,x2):
    if len(x1)!=len(x2):
        print(f'Różne długości listy wektorów i listy skalarów: {len(x1)} i {len(x2)}')
        return "error"
    x=[]
    for i,j in zip(x1,x2):
        x.append(multi_by_skalar(i,j))
    y=[x1[0][0]]
    for i in range(1,len(x1[0])):
        z=0
        for j in x:
           z=z+j[i]
        y.append(z)
    return(y)
```
